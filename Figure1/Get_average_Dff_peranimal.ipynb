```python
amplitude = {k:[] for k in dff.taskstoplot + ['Task2a', 'Task2b']}
for a in dff.animalname:
    amp = np.load(os.path.join(dff.Foldername, a,
                                     'SaveAnalysed', 'transient_properties.npz'), allow_pickle=True)['amplitude'].item()
    for t in dff.taskstoplot + ['Task2a', 'Task2b']:
        amplitude[t].extend(amp[t])
```

```python
fs, ax = plt.subplots(1, 2, dpi=200, figsize=(10, 4))
for t in ['Task1', 'Task2a', 'Task2b', 'Task3']:
    sns.distplot(amplitude[t], ax=ax[0])
    ax[1].hist(amplitude[t], bins=np.linspace(0, 1, 1000), alpha=0.5,
               histtype='step', cumulative=True, density=True, label=t)
    ax[1].set_xlim((0.1, 0.99))
    ax[0].set_xlim((0, 3))
ax[1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
